```python
maker_stats_tmp = open("/Users/dshipilina/GitHub/GenomeVanessa/VanessaAnnotation/maker3run2_m.all.maker.proteins.maker.stats.out").read()
maker_stats_tmpp = maker_stats_tmp.split('\n')
maker_stats_tmpp_array=[]
#print(split_file2[2])
for i in maker_stats_tmpp:
    ii=i.split()
    maker_stats_tmpp_array+=[ii]

MAKER_stat = pd.DataFrame(maker_stats_tmpp_array, columns = ["Length of the 5 UTR", \
                          "Fraction of splice sites, EST/mRNA-seq confirmed", \
                          "Fraction of exons that match an EST/mRNA-seq alignment",\
                          "Fraction of exons that overlap EST/mRNA-seq or protein alignments",\
                          "Fraction of splice sites confirmed by ab initio gene prediction",\
                          "Fraction of exons that overlap an ab initio gene prediction",\
                          "Number of exons in the mRNA",\
                          "Length of the 3 UTR","Length of the protein sequence produced by the mRNA"])
MAKER_stat = MAKER_stat.apply(pd.to_numeric, errors='coerce', axis=1)
MAKER_stat= MAKER_stat.drop(labels=[7982], axis=0)
MAKER_stat

ml.rcParams['figure.figsize'] = (5.0, 5.0)
sns.distplot(MAKER_stat["Length of the 5 UTR"],bins=50)
plt.xlabel("Length of the 5 UTR")
plt.ylabel("Density (KDE)")
plt.savefig("/Users/dshipilina/GitHub/GenomeVanessa/VanessaAnnotation/5primeUTR_lenght.png",bbox_inches="tight")
plt.show()

ml.rcParams['figure.figsize'] = (5.0, 5.0)
sns.distplot(MAKER_stat["Length of the 3 UTR"],bins=50)
plt.xlabel("Length of the 3 UTR")
plt.ylabel("Density (KDE)") 
plt.savefig("/Users/dshipilina/GitHub/GenomeVanessa/VanessaAnnotation/3primeUTR_lenght.png",bbox_inches="tight")
```
